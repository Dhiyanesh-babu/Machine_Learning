```python
# Importing the datsets

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score,mean_absolute_error,mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
```

```python
# Preprocessing the data

data =pd.read_csv('dataset.csv')
for i in range(len(data['Gender'])):
    if data['Gender'][i]=="Male":
        data['Gender'][i]=1
    else:
        data['Gender'][i]=0
```

```python
data.head()
```

```python
# Splitting the datasets

x=data.drop(columns='Personality (Class label)')
print(x)
y=data['Personality (Class label)']
print(y)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.05,random_state=42)
```

```python
# Decison Tree algorithm

clf=DecisionTreeClassifier(criterion='gini',min_samples_split=2)
clf.fit(x_train,y_train)
```

```python
tree.plot_tree(clf)
```

```python
y_pred=clf.predict(x_test)
```
